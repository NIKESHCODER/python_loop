```python
#Q10. Write a program to display the number names of the digits of a number entered by user, 
#for example if the number is 231 
#then output should b e 
#Two
#Three
#Oneĺ
num = input("Enter a number: ")
digit_names = {
    '0': 'Zero',
    '1': 'One',
    '2': 'Two',
    '3': 'Three',
    '4': 'Four',
    '5': 'Five',
    '6': 'Six',
    '7': 'Seven',
    '8': 'Eight',
    '9': 'Nine'
}
print("\nNumber Names:")
for digit in num:
    if digit in digit_names:
        print(digit_names[digit])
    else:
        print("Invalid input! Please enter only digits.")
        break

```
